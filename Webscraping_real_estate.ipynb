```
#import packages 
import requests
from bs4 import BeautifulSoup
```

```
url = 'https://www.makaan.com/hyderabad-residential-property/buy-property-in-hyderabad-city'
response = requests.get(url)
print(response)
```

```
print(response.text)
```

```
soup = BeautifulSoup(response.text)
print(soup)
```

```
single_li = soup.find('li',class_= 'cardholder')
```

```
single_li.find('div',class_='second-line').text
```

```
single_li.find('span',attrs={'class' : 'val','itemprop' : "offers"}).text
```

```
single_li.find('span',class_='val').text
```

```
single_li.find('td',class_='size').text
```

```
single_li.find('td',class_='val').text
```

```
all_lst = soup.find_all('li',class_='cardholder')
owner_type = []
price = []
bhk = []
area = []
con_status = []
for lst in all_lst:
  if lst.find('div',class_='second-line') is not None:
    a = lst.find('div',class_='second-line').text
    
    owner_type.append(a)

    b = lst.find('span',attrs={'class' : 'val','itemprop' : "offers"}).text
    price.append(b)

    c = lst.find('span',class_='val').text
    bhk.append(c)

    d = lst.find('td',class_='size').text
    area.append(d)

    e = lst.find('td',class_='val').text
    con_status.append(e)
```

```
print(len(owner_type))
print(len(area))
```

```
import pandas as pd
data = pd.DataFrame({'Owner' : owner_type,
                     'Price' : price,
                     'BHK' : bhk,
                     'Area' : area,
                     'Status' : con_status})
data
```

```
data.to_csv('real-state')
```
